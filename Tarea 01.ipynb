3.- Escribir una funcion que reciba como argumentos el numero de horas y el costo por hora dejarlo
como argumento predefinido en 150 la funcion debe de regresar el pago del empleado en un
dıa, en una semana (5 dıas) y en un mes (20 dıas).

```
# 1 dia = 24 horas
# 1 semana = 120 horas
# 1 mes = 480 horas
def Empleado(horas,costo=150):
  pago=horas*costo
  dias=horas/24
  semanas=dias/5
  meses=dias/20
  if horas>=24 and horas<120:
    print("Dias de trabajo: ",dias)
  elif horas>=120 and horas<480:
    print("Semanas de trabajo: ",semanas)
  elif horas>=480:
    print("Meses de trabajo: ",meses)
  
  print("El pago total es de: ")
  return pago

c=float(input("Ingresa las horas de trabajo: "))
Empleado(c)
```

5.- Una tienda de juguetes tiene dos productos estrella, un payaso y un juego de mesa, el payaso
pesa 150 gr y el juego de mesa pesa 170 gr. Hacer una funcion que reciba el nuermo de payasos
y el numero de juegos de mesa por pedido, si el peso total del pedido es mayor a 10 kg, debe de mandar error, debe de inidicar por cuantos gramos se paso el envıo y debe de regresar False
al programa principal.

```
#Cantidad maxima de pedidos de payasos:66
#Cantidad maxima de pedidos de juegos de mesa:58
def Pedido(cantidad,payaso=150,jmesa=170):
  ps=payaso
  jm=jmesa
  pro=input("Ingrese el nombre del producto: ")
  if pro=='ps' or pro=='Ps':
    peso=cantidad*payaso
    print("Peso Total del envio: ",peso)
  elif pro=='jm' or pro=='Jm':
    peso=cantidad*jmesa
    print("Peso Total del envio: ",peso)
  if peso>10000:
    exceso=peso-10000
    print("ERROR:Exceso de peso")
    print("Cantidad de peso extra: ",exceso)
    return False
ped=int(input("Ingrese la cantidad del pedido: "))
Pedido(ped)
```

```
def Salario(salario):
  if salario<10000:
    impuesto=0.05
    paga=salario*impuesto
    total=salario-paga
    print("Tu salario es de menos de 10 mil")
    print("Total de impuestos: ",paga)
    print("Tu paga libre de impuestos es: ",total)
  elif salario>=10000 and salario <15000:
    impuesto=0.1
    paga=salario*impuesto
    total=salario-paga
    print("Tu salario esta entre 10 mil y 15 mil")
    print("Total de impuestos: ",paga)
    print("Tu paga libre de impuestos es: ",total)
  elif salario>=15000 and salario <20000:
    impuesto=0.15
    paga=salario*impuesto
    total=salario-paga
    print("Tu salario esta entre 15 mil y 20 mil")
    print("Total de impuestos: ",paga)
    print("Tu paga libre de impuestos es: ",total)
  elif salario>=20000:
    impuesto=0.18
    paga=salario*impuesto
    total=salario-paga
    print("Tu salario es de mas de 20 mil")
    print("Total de impuestos: ",paga)
    print("Tu paga libre de impuestos es: ",total)
  return True

sal=float(input("Ingresa el salario: "))
Salario(sal)
```

8.- Escribir un programa que pida al usuario un numero entero y muestre en pantalla un trıangulo rectangulo

```
def triangulo(n):
    return "\n".join(["*"*i for i in range(1,n+1)])
numero=int(input("indica un numero: "))
print(triangulo(numero))
```

9.- Escribir una funcion que reciba un numero entero positivo y devuelva su factorial


```
import math
def Factorial(f):
  if f==0:
    print("Factorial = 1")
  elif f>0:
    f=math.factorial(f)
    print("Valor de factorial: ")
    return f

fac=int(input("Ingresa un numero: "))
Factorial(fac)
```

10.-Escribir una funcion que calcule el area de un cırculo y ota que calcule el volumen de un cilindro
utilizando la primera funcion.

```
import math
def Area(radio):
  pi=math.pi
  area=pi*(radio**2)
  print("Area del circulo:", area)
def Volumen(radio,altura):
  pi=math.pi
  area=pi*(radio**2)
  volumen=area*altura
  print("Volumen del cilindro: ")
  return volumen
ra=float(input("Ingrese el valor del radio: "))
vol=float(input("Ingrese el valor de la altura del cilindro: "))
Area(ra)
Volumen(ra,vol)
```
